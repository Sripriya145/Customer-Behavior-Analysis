```python
#group the numeric column
numeric_col=df.select_dtypes(include="number")

#find the mean
mean_value=numeric_col.mean()

#find the median
median_value=numeric_col.median()

#find the sum
sum_value=numeric_col.sum()

#display all content
print("Mean")
print(mean_value)
print("\nMedian")
print(median_value)
print("\nSum")
print(sum_value)
```

```python
#find the range value
range_value=numeric_col.max() -numeric_col.min()

#find the standard deviation
std_value=numeric_col.std()

#find the variance
variance_value=numeric_col.var()

#find the iqr
iqr_value1=numeric_col.quantile(0.25)
iqr_value3=numeric_col.quantile(0.75)
IQR=iqr_value3-iqr_value1

#find the correlation
co_relation_value=numeric_col.corr()

#group the non-numeric colunms
ColValue=df.select_dtypes(include=["number","object"])

#find the mode
mode_value=ColValue.mode()

#display all content
print("Range Value")
print(range_value)
print("\nVariance")
print(variance_value)
print("\nIQR")
print(IQR)
print("\nStandard deviation")
print(std_value)
print("\nCorelational")
print(co_relation_value)
print("\nMode")
print(mode_value)
```

```python
# group by GRP column for fur usage
group_data_grp=df.groupby("GRP")

def operation_fun():
    agger_operation={
        'BILL_AMT':['sum','mean','median'],
        'VALUE':['sum','mean','median'],
        'PRICE':['sum','mean','median']
    }
    return agger_operation;

#Grouping data and performing aggregation
res_group_grp  =  group_data_grp[['BILL_AMT','VALUE','PRICE']].agg(operation_fun());

#print grouped result
print(res_group_grp)

group_data_sgrp=df.groupby("SGRP")



```
